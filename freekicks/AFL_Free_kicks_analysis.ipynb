# Analysis of Free kicks awarded in home and away games for all seasons
----

Do home teams get more free kicks than their visiting opponents?

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
```

## Load the csv created from data obtained from https://afltables.com/afl/afl_index.html

```python
# File to Load
aflfreekicks_df = pd.read_csv("freekicks-all-time-home-away.csv")
aflfreekicks_df.columns
aflfreekicks_df.head(3)
```

### Create separate data frames for home games and away games to make plotting easier

```python
afl_freekicks_hometeam_df=aflfreekicks_df[["Team", "freekicks_for_home_team","freekicks_against_home_team"]]
afl_freekicks_awayteam_df=aflfreekicks_df[["Team", "freekicks_for_away_team","freekicks_against_away_team"]]
afl_freekicks_hometeam_df.columns
```

### Plot freekicks awarded for and against teams playng at HOME

```python
labels = afl_freekicks_hometeam_df["Team"]
freekicks_for = afl_freekicks_hometeam_df["freekicks_for_home_team"]
freekicks_against = afl_freekicks_hometeam_df["freekicks_against_home_team"]

plt.rcParams["figure.figsize"] = [10,10]
plt.rcParams["figure.autolayout"] = True
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.barh(x + width/2,freekicks_for, width, label='Free kicks For Home Team', facecolor="red",)
rects2 = ax.barh(x - width/2, freekicks_against, width, label='Free kicks Aganst Home Team',facecolor="blue",)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Teams')
ax.set_title('All time Freekicks awarded playing at HOME')
ax.set_xlabel('Freekicks Home and Away')
ax.set_yticks(x)
ax.set_yticklabels( labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

plt.savefig("Freekicks_HOME.jpg")
plt.show()
```

### Plot freekicks awarded for and against teams playng AWAY

```python
abels = afl_freekicks_awayteam_df["Team"]
freekicks_for = afl_freekicks_awayteam_df["freekicks_for_away_team"]
freekicks_against = afl_freekicks_awayteam_df["freekicks_against_away_team"]

plt.rcParams["figure.figsize"] = [10,10]
plt.rcParams["figure.autolayout"] = True
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.barh(x + width/2,freekicks_for, width, label='Free kicks For Away Team', facecolor="red",)
rects2 = ax.barh(x - width/2, freekicks_against, width, label='Free kicks Aganst Away Team',facecolor="blue",)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Teams')
ax.set_title('All time Freekicks awarded playing AWAY')
ax.set_xlabel('Freekicks Home and Away')
ax.set_yticks(x)
ax.set_yticklabels( labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

plt.savefig("Freekicks_AWAY.jpg")
plt.show()
```

### The conclusion that may be drawn from the two graphs is that teams playing at home do tend to get more freekicks in their favour whereas laying away gets them less. There are some exceptions but that appears to be with Melbourne based clubs who do play a lot of their away games at their home ground but with less of their fans able to attend. So fan support appears to have some bearing on the outcome.
