# Analysis of wins and losses for each team whilst playing at HOME for all seasons
----

Do teams do worse when playing away from home and in front of less of their fans?

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
```

### Load the csv created from data obtained from https://afltables.com/afl/afl_index.html

```python
# File to Load
afl_winloss_home_df = pd.read_csv("winloss_home.csv")
afl_winloss_home_df.columns
afl_winloss_home_df.head()
```

### Plot of wins vs losses for teams playing at Home


```python
labels = afl_winloss_home_df["Team"]
wins = afl_winloss_home_df["Won"]
losses = afl_winloss_home_df["Loss"]

plt.rcParams["figure.figsize"] = [10,10]
plt.rcParams["figure.autolayout"] = True
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.barh(x + width/2,wins, width, label='Wins', facecolor="red",)
rects2 = ax.barh(x - width/2, losses, width, label='Losses',facecolor="blue",)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Teams')
ax.set_title('All time wins vs losses playing at Home')
ax.set_xlabel('Wins vs Losses')
ax.set_yticks(x)
ax.set_yticklabels( labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)


plt.show()
```

## There is clearly a trend that teams playing at HOME win more games than when they play AWAY. There are some exceptions for example Brisbane Bears, they were a new team established in NRL territory so they had few fans attending their home games but when they became the Brisbane Lions after merging with the established Melbourne club Fitzroy they gained a lot more fans and their win/loss record went as expected.

## The difference in win/loss records for Melbourne based teams is clear. They have a larger fan base and they don't have to travel interstate as much so more of their fans can attend AWAY games.
