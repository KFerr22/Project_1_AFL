```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
avg_scores_csv = "Average Scores (2000-2021).csv"
Average_Scores = pd.read_csv(avg_scores_csv)

Average_Scores 
```

```python
y_axis = Average_Scores['Average Scores']
x_axis = Average_Scores['Year']

plt.bar(x_axis, y_axis)
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis , rotation=90)
plt.xlabel('Year')
plt.ylabel('Teams Average Final Score')
plt.title('Average Final Score by Season\n*2020 Shortened Quarter Lengths (COVID-19)')

plt.savefig('Average Final Score by Season.png')
plt.show()
```

What does this Data Show?

- Since the year 2000 the Average Final Score has been on a steady decline 

- 2020 is an oulier because of the reduced quarter lengths because of the Covid - 19 affected season

- Why would this matter for the AFL? The AFL wants a higher scoring game for greater fan engagement and to produce a better product for enternaimnet purposes and TV. 

- To secure a better TV rights deal the AFL wants more goals kicked which in turn leads to more Advert breaks which will increase revenue

```python
avg_stats_csv = "Team Averages (2000-2021).csv"
Yearly_Average_Statistics = pd.read_csv(avg_stats_csv)

Yearly_Average_Statistics
```

```python
Full_Stats_DF = pd.merge(Average_Scores, Yearly_Average_Statistics, how= 'outer', on= 'Year')

Full_Stats_DF = Full_Stats_DF.drop(20)

Full_Stats_DF
```

```python
x_value = Full_Stats_DF['Tackles']
y_value = Full_Stats_DF['Average Scores']
correlation = st.pearsonr(x_value, y_value)

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
regress_values = x_value * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_value, y_value)
plt.plot(x_value,regress_values,'r-')
plt.xlabel('Average Tackles per Game')
plt.ylabel('Teams Average Final Score ')
plt.title('Teams Average Final Score by Average Tackles per Game\n* Outlier = 2020 Shortened Quarter Lengths (COVID)')
plt.annotate((f'''The correlation between\nvalues is {round(correlation[0],2)}'''),(35,81),fontsize=15,color='red')

print(f'''The correlation between values is {round(correlation[0],2)}''')
plt.savefig('Teams Average Final Score by Average Tackles per Game.png')
plt.show()
```
